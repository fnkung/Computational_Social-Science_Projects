```python
from transformers import pipeline, set_seed
# Here, 'text-generation' tells the pipeline to create a text generation model, and model='gpt2'
# specifies that you want to use the GPT-2 model for this task.
generator = pipeline('text-generation', model='gpt2')
```

```python
set_seed(100)
input_text = "This project uses natural language processing techniques to explore the tweets of members of the 116th Congress and questions of political polarization. Through this analysis, we found "
# specifying the maximum length of the output, and the number of different sequences you want
output = generator(input_text, max_length=150, num_return_sequences=5, temperature = 0.5)
for result in output:
    print(result['generated_text'])
    print("----")
```
